**standardization**

- standardization means scalining the data into the scale
- we have different columns units so that the value will vary
- one column has very huge values
- another column has very less value
- so it is important to scale all types of units under one scale
- we have 2 procedures
- standardization
  - Z-score
    - the values ranges from -3 to 3

**normalization**

- min max scalar
- values ranges from 0 to 1

```python
# take the prevailing wage column
# Z-score = x-mean/sigma

# step-2:calculate mean of prevaling wage
# step-3:calculate std of prewage
# step-4:Nr pwage-mean
# step-5:pwage_Zscore=Nr/Dr
```

```python
pwage=visa_df['prevailing_wage']
pwage_mean=visa_df['prevailing_wage'].mean()
pwage_std=visa_df['prevailing_wage'].std()
Nr=pwage-pwage_mean
visa_df['prevailing_wage_z']=Nr/pwage_std
```

**normalization**

$minmaxscalar$

```python
# x-x min/(x_max-x_min)
# x=data column
# x_min
# x_max
# nr=x-x_min
# dr=x_max-x_min
# nr/dr
```

```python
x=visa_df['prevailing_wage']
x_min=visa_df['prevailing_wage'].min()
x_max=visa_df['prevailing_wage'].max()
nr=x-x_min
dr=x_max-x_min
visa_df['prevailing_wage_norm']=nr/dr
visa_df['prevailing_wage_norm']
```

$Min Max-Scalar$

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
visa_df['prevailing_wage_mms']=mms.fit_transform(visa_df[['prevailing_wage']])
visa_df
```
